```python
from collections import defaultdict
import requests

import pandas as pd
```

```python
url = "https://itunes.apple.com/gb/rss/customerreviews/id=1500780518/sortBy=mostRecent/json"
r = requests.get(url)
```

```python
r.status_code
```

```python
r.json()["feed"]["entry"][30]
```

```python
data = r.json()["feed"]["entry"]
len(data)
```

```python
# try to read raw json with pandas
from io import StringIO

pd.read_json(StringIO(r.content.decode("utf-8")))
```

```python
# try to read raw dict with pandas
pd.DataFrame(data).head()
```

```python
parsed_data = defaultdict(list)

for entry in data:
    for key, val in entry.items():
        for subkey, subval in val.items():
            # take all fields with label as column
            if subkey == "label":
                parsed_data[key].append(subval)
            
            # if field equals "attributes", take all elements in it as columns
            if subkey == "attributes":
                for att_key, att_val in subval.items():
                    parsed_data[f"{key}_{att_key}"].append(att_val)
                    
```

```python
df = pd.DataFrame(parsed_data)
```
